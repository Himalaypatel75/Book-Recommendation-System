## Popularity based recommendation system

```python
#merging rating and book with ISBN
rating_book = rating.merge(book, on = 'ISBN')
```

```python
rating_book.head()
```

```python
rating_book.shape
```

```python
top_ten_rating_books = rating_book.copy().groupby('Book-Title').count().sort_values('Book-Rating', ascending=False)
```

```python
top_ten_rating_books.head(10)
```

```python
# Ensure the 'Book-Rating' column is numeric
rating_book['Book-Rating'] = pd.to_numeric(rating_book['Book-Rating'], errors='coerce')

# Drop rows with NaN values in 'Book-Rating'
rating_book = rating_book.dropna(subset=['Book-Rating'])

# Now, calculate the mean and sort the top 10 books by rating
average_ten_rating_books = (
    rating_book.groupby('Book-Title')['Book-Rating']
    .mean()
    .sort_values(ascending=False)
)

print(average_ten_rating_books)
```

```python
popularity_df = top_ten_rating_books.merge(average_ten_rating_books, on='Book-Title')
```
